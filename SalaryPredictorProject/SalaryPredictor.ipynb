```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


w = 0
b = 0

df = pd.read_csv('Salary_Data.csv')
x = df["YearsExperience"].values
y = df["Salary"].values

learningrate = 0.0001
epochs = 1000000
n = len(x)
for i in range(epochs):
    y_pred = w*x + b

    error = y - y_pred
    mse = np.sum(error**2)

    dw = -2/n * np.sum(x* error)
    db = -2/n * np.sum(error)
    
    w -= dw * learningrate
    b -= db * learningrate
    
    
    if i % 1000 == 0:
        print(f"Epoch {i}: mse={mse:.2f}, w={w:.2f}, b={b:.2f}")
        plt.scatter(x, y, color="blue", label="Actual")
        plt.plot(x, y_pred, color="red", label="Predicted")
        plt.xlabel("Years of Experience")
        plt.ylabel("Salary")
        plt.legend()
        plt.show()


y_pred = w*x + b

mse = np.mean((y - y_pred)**2)


print("\nFinal Metrics:")
print(f"MSE: {mse:.2f}")

# Plot data and regression line
plt.scatter(x, y, color="blue", label="Actual")
plt.plot(x, y_pred, color="red", label="Predicted")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()

# Prediction function
def predict_salary(years):
    return w * years + b

print(f"Predicted salary for 3.5 years: {predict_salary(3.5):.2f}")
print(f"Predicted salary for 7 years: {predict_salary(7):.2f}")


mape = np.mean(np.abs((y - y_pred)/y)) * 100
accuracy = 100 - mape
print(f"Prediction Accuracy: {accuracy:.2f}%")

```
